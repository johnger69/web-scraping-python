```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.expansion.com/mercados/cotizaciones/indices/ibex35_I.IB.html'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content,'html.parser')
```

```python
soup
```

```python
tabla = soup.find('table',attrs={'id':'listado_valores'})
tabla
```

```python
tabla = soup.find('table',attrs={'id':'listado_valores'}).find('tbody')
tabla
```

```python
filas = soup.find('table',attrs={'id':'listado_valores'}).find('tbody').find_all('tr')
filas[0]
```

```python
filas[0].find_all('td')[0].get_text()
```

```python
filas[0].find_all('td')[0].text
```

```python
nombres=[]
for fila in filas:
  nombres.append(fila.find_all('td')[0].get_text())
```

```python
nombres
```

```python
valores=[]
for fila in filas:
  valores.append(fila.find_all('td')[1].get_text())
```

```python
valores
```

```python
df = pd.DataFrame({'Nombres':nombres,'Valores':valores})
```

```python
df
```

```python
df.to_csv('valoresBolsa.csv')
```

#Obteniendo imagenes

```python
import requests
from bs4 import BeautifulSoup
```

```python
url= 'https://www.josecodetech.es'
web = requests.get(url)
soup = BeautifulSoup(web.text,'lxml')
```

```python
soup
```

```python
tarjeta = soup.find('div',attrs={'class':'card'})
imagen = tarjeta.find('img').get('src')
imagen = imagen [1:len(imagen)]
imagen
```

```python
enlace_imagen = url + imagen
print(enlace_imagen)
```

```python
if len(enlace_imagen)==0:
  print("No encuentro la imagen")
else:
  print(enlace_imagen)
```

```python
from IPython.display import Image
```

```python
# mostramos la imagen obtenida
Image(enlace_imagen)
```

```python
# Para todas las imagenes
tarjetas = soup.find_all('div',attrs={'class':'card'})
#print(tarjetas)
enlaces_imagenes=[]
for tarjeta in tarjetas:
  imagen = tarjeta.find('img').get('src')
  imagen = imagen[1:len(imagen)]
  enlace_imagen = url + imagen
  enlaces_imagenes.append(enlace_imagen)
```

#Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.josecodetech.es/'
```

```python
web = requests.get(url)
```

```python
web.status_code
```

```python
print(web.text)
```

```python
# obtenemos con bs4
soup = BeautifulSoup(web.text,'lxml')
soup
```

```python
# obtenemos primer valor con etiqueta ul
soup.find('ul')
```

```python
enlaces = soup.find('ul').find_all('a')
enlaces
```

```python
# Buscar clase, por atributo y valor
soup.find('div',attrs={'class':'card'})
```

```python
# volviendo con los enlaces
enlace = enlaces[0]
enlace
```

```python
enlace.get('href')
```

```python
enlace.text
```

```python
enlace.get_text()
```

```python
# para todos los enlaces obtener el link
links = [enlace.get('href') for enlace in enlaces]
links
```

```python
# nuevas peticiones url
nueva_url = links[1]
nueva_web = requests.get(nueva_url)
nueva_web.status_code
```

```python
soup_nueva = BeautifulSoup(nueva_web.text,'lxml')
print(soup_nueva)
```
